```python
cat_cols = [
    'schoolsup','famsup','paid','activities',
    'nursery','higher','internet','romantic','passed'
]

for col in cat_cols:
    df[col]=df[col].str.strip().str.capitalize()
```

```python
df[cat_cols].head()
```

```python
df.head()
```

```python
df = df[(df['age']>10)&(df['age']<25)]
```

```python
df=df[df['failures']>=0]
```

```python
df=df[(df['absences']>=0)&(df['absences']<=30)]
```

```python
df.describe()
df.info()
```

```python
df=df[(df['studytime']>=1&(df['studytime']<=4))]
```

```python
df.head()
```

```python
df.to_csv('student_cleandata.csv', index=False)
```
